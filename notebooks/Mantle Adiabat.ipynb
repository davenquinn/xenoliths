The mantle adiabatic temperature gradient is governed by the equation $$\left ( \frac{\partial T}{\partial z} \right )_S = \frac{agT}{C_p}$$ 

- $T$: temperature
- $z$: depth
- $g$: gravitational acceleration
- $α$: thermal expansion coefficient of material
- $C_p$: isobaric heat capacity of material

Assume $\left ( \frac{\partial T}{\partial z} \right )_S = 0.4$ K km$^{-1}$

```python
from geotherm.units import u
import numpy as N

dTdz = u(0.3,"K/km") # for greater depths
a = u(3e-5,"1/K")
Cp = u(1,"kJ/kg")
T = u(1600,"K")
g = u(10,"m*s**-2")

z = N.linspace(0, u(400,"km"), 1000)

def adiabat(z, tlim):
    return a*g*tlim[1]/Cp*z

res = adiabat(z,[0,T])
```
