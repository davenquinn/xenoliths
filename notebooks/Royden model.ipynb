```python
depths = N.linspace(0,150,150)*1e3

vanilla = RoydenModel()

royden = RoydenModel(
    Kl=3.35,
    Al=1e-15,
    zr=30e3)

profile = vanilla(50e3,depths,30e3)

fig, ax = plt.subplots()
ax.plot(profile, depths, label="Vanilla")

for d in N.linspace(0,60,5):
    ax.plot(royden(d*1e3, depths, d/2*1e3), depths, label="Model-{0:.0f}".format(d))

ax.invert_yaxis()
ax.legend()
```
